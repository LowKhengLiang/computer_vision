```python
import os

import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline

from zipfile import ZipFile
from urllib.request import urlretrieve
```

```python
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
```

```python
URL = r"https://www.dropbox.com/s/qhhlqcica1nvtaw/opencv_bootcamp_assets_NB1.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB1.zip")

# Download if assest ZIP does not exists.
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path)
```

### Splitting and Merging Color Channels

cv2.split(): Divides a multi-channel array into several single-channel arrays

cv2.merge(): Merges several arrays to make a single multi-channel array. All the input matrices must hae the same size.

```python
img_NZ_bgr = cv2.imread("New_Zealand_Lake.jpg", cv2.IMREAD_COLOR) #Read in as a color image
b,g,r = cv2.split(img_NZ_bgr) #bgr because cv2 reads in bgr

#Show the channels
plt.figure(figsize=[20, 5])
plt.subplot(141);plt.imshow(r, cmap="gray");plt.title("Red Channel")
plt.subplot(142);plt.imshow(g, cmap="gray");plt.title("Green Channel")
plt.subplot(143);plt.imshow(b, cmap="gray");plt.title("Blue Channel")

#Merge the individual channels into a BGR image
imgMerged = cv2.merge((b,g,r))
plt.subplot(144);plt.imshow(imgMerged[:, :, ::-1]); plt.title("Merged Output")

#Red is darker because there are less red in the image
```

### Converting to different Color Spaces

cv2.cvtColor() converts an image from one color space to another.

dst = cv2.cvtColor(src, code)

dst: Is the output image of the same size and depth as src.

src: input image: 8-bit unsigned, 16-bit unsigned, or single-precision floating-point

code: color space conversion code (see ColorConversionCodes)

```python
img_NZ_rgb = cv2.cvtColor(img_NZ_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_NZ_rgb)
```

### Changing to Hue Saturation Value color space

```python
img_hsv = cv2.cvtColor(img_NZ_bgr, cv2.COLOR_BGR2HSV)
#split image into BGR components
h,s,v = cv2.split(img_hsv)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title('Hue Channel');
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title('Saturation Channel');
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title('Value Channel');
plt.subplot(144);plt.imshow(img_NZ_rgb);plt.title("Original");
plt.show()

#hue = color
#saturation = intensity of color
#value = how light or dark
```

### Modifying Individual Channel

```python
h_new = h+10
img_NZ_merged = cv2.merge((h_new,s,v))
img_NZ_bgr_new = cv2.cvtColor(img_NZ_merged, cv2.COLOR_HSV2RGB)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title('H Channel');
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title('S Channel');
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title('V Channel');
plt.subplot(144);plt.imshow(img_NZ_bgr_new);plt.title("Modified");
plt.show()
```

### Saving Images

```python
cv2.imwrite('New_Zealand_Lake_SAVED.png', img_NZ_bgr_new)
Image(filename='New_Zealand_Lake_SAVED.png')
```
